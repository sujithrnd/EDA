```python
import matplotlib.pyplot as plt
category_features=[col for col in data1.columns if data1[col].dtype=='O']
category_features
```

```python
# categorical columns
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(category_features)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=data1[category_features[i]],palette="Set2")
    plt.xlabel(category_features[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=data1[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```
