```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
numeric_features=[ 'having_IP_Address',  'SFH',  'Redirect',   'DNSRecord', 'web_traffic', 'Statistical_report']

plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```

```python
#univariate analysis

#most popular category in app


df["Abnormal_URL"].value_counts().plot.pie(y=df["Abnormal_URL"],figsize=(15,15),autopct='%1.1f%%')
```

```python
df_cat=df["having_IP_Address"].value_counts()[:20]
category = pd.DataFrame(df['having_IP_Address'].value_counts())        #Dataframe of apps on the basis of DNSRecord
category.rename(columns = {'having_IP_Address':'Count'},inplace=True)
plt.figure(figsize=(15,6))
sns.barplot(x=df_cat, y ='Count',data = category[:10],palette='hls')
plt.title('Top 10 App categories')
plt.xticks(rotation=90)
plt.show()
```
